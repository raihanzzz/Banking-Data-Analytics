```python
#examine the distribution of unique categories
categorical_cols = df[["BRId","IAId","GenderId","Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Income Band"]].columns

for col in categorical_cols:
    print(f"Value Counts for '{col}':")
    display(df[col].value_counts())
```

## univarient analysis

```python
for i, predictor in enumerate(df[["BRId","IAId","GenderId","Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Income Band"]].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predictor)
```

## bivarient analysis

```python
for i, predictor in enumerate(df[["BRId","IAId","GenderId","Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Income Band"]].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue='GenderId')
```

```python
for i, predictor in enumerate(df[["BRId","IAId","GenderId","Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Income Band"]].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue='Nationality')
```

```python
# histplot of values of different occupation
for col in categorical_cols:
    if col == "Occupation":
        continue
    plt.figure(figsize=(8,4))
    sns.histplot(df[col])
    plt.title('Histogram of occupation count')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

## Numerical Analysis

```python
numerical_cols = ['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']

plt.figure(figsize=(15,20))
for i, col in enumerate(numerical_cols):
    plt.subplot(4,3,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)
plt.show()
```

## Heat Maps

```python
numerical_cols = ['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']

corellation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(9,9))
sns.heatmap(corellation_matrix, annot=True, cmap='crest', fmt=".2f")
plt.title("Corellation Matrix")
plt.show()
```
